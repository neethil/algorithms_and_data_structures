In this example, 4 is the root, 2 and 6 are the left and right children of 4. 1, 3, 5, 7 are the children of 2 and 6. The output returns a list of lists where each list consists of node values in the order they appear in the path from the root to the leaf of the branch.


-   Copy the solution your partner wrote. 


```python
# Your answer here

from typing import List

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def bt_path(root: TreeNode) -> List[List[int]]:
    if root is None:
        return []

    result = []

    def dfs(node: TreeNode, path: List[int]):
        if node is None:
            return

        # we add the current node to the path
        path.append(node.val)

        # if it's a leaf node, add the path to the result
        if node.left is None and node.right is None:
            result.append(path[:])  # Append a copy of the path

        # Recur for left and right children
        dfs(node.left, path)
        dfs(node.right, path)

        # Backtrack
        path.pop()

    dfs(root, [])
    return result
```


-   Explain why their solution works in your own words.


```python
# Your answer here
```

This solution works because it uses depth-first search (DFS) to explore all paths from root to leaf. At each node, the function dfs recursively checks for the left and right children and builds a path from root to the current node. Also for each node, the value of current node is stored to path. If current node has no children, that means we have reached a leaf and the path is recorded as a root to leaf path and this complete path is appended to the result list. After the function has traversed the left and right subtrees of a node, it also uses backtracking to remove the current node from the path using the pop method. This ensures that the path list is reset when the function returns to the parent node to explore the other branches, to remove values from previously explored branches.
